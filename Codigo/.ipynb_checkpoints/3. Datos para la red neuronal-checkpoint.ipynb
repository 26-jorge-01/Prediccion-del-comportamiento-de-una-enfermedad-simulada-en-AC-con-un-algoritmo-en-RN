## Modelo SIS simple

```python
a=0.2; b=0.5
system = initialCondition(0.1, np.zeros((10,10)))
SISmodel = models("sis",a,b,system,Moore)
```

```python
SISmodelApplied = SISmodel.basicModel(10)

df = pd.DataFrame()
x,y = system.shape
n = x*y

df["# Susceptibles"] = [int(n*m) for m in SISmodelApplied[1][0]]
df["# Infectados"] = [int(n*m) for m in SISmodelApplied[1][1]]
df["% Susceptibles"] = SISmodelApplied[1][0]
df["% Infectados"] = SISmodelApplied[1][1]

df.head()
```

```python
SISmodelEvolutions = SISmodel.evolutions
listOfCoordinatesNeighbors = Neighbors(system,Moore,1,1)

df = GetStatesByIteration(SISmodelEvolutions,listOfCoordinatesNeighbors).head()
df
```

```python
NeighborsStatesInDataFrame([1,6],listOfCoordinatesNeighbors,df).head()
```

## Modelos con Natalidad y mortalidad

```python
a=0.2; b=0.5
ranges = [[1,25,0.4],[25,64,0.4],[65,90,0.2]]
ages_sys1 = agesMatrix(ranges,system)
mr = [[1,14,0.0005],[15,69,0.02],[70,100,0.15]]
br = 2

system = initialCondition(0.1, np.zeros((10,10)))
SIRmodelWithBirthsAndDeaths = models("sir_birthAndMortavility",a,b,system,Moore)
SIRmodelWithBirthsAndDeaths.birthRate = br; SIRmodelWithBirthsAndDeaths.probabilityOfDyingByAgeGroup = mr
SIRmodelWithBirthsAndDeaths.systemAges = ages_sys1; SIRmodelWithBirthsAndDeaths.annualUnit = 365
```

```python
SIRmodelWithBirthsAndDeathsApplied = SIRmodelWithBirthsAndDeaths.basicModel(10)

df = pd.DataFrame()
x,y = system.shape
n = x*y

df["# Susceptibles"] = [int(n*m) for m in SIRmodelWithBirthsAndDeathsApplied[1][0]]
df["# Infectados"] = [int(n*m) for m in SIRmodelWithBirthsAndDeathsApplied[1][1]]
df["# Recuperados"] = [int(n*m) for m in SIRmodelWithBirthsAndDeathsApplied[1][2]]
df["# Muertos"] = [int(n*m) for m in SIRmodelWithBirthsAndDeathsApplied[1][3]]
df["% Susceptibles"] = SIRmodelWithBirthsAndDeathsApplied[1][0]
df["% Infectados"] = SIRmodelWithBirthsAndDeathsApplied[1][1]
df["% Recuperados"] = SIRmodelWithBirthsAndDeathsApplied[1][2]
df["% Muertos"] = SIRmodelWithBirthsAndDeathsApplied[1][3]

df.head()
```

```python
SIRmodelWithBirthsAndDeathsEvolutions = SIRmodelWithBirthsAndDeaths.evolutions
listOfCoordinatesNeighbors = Neighbors(system,Moore,1,1)

df = GetStatesByIteration(SIRmodelWithBirthsAndDeathsEvolutions,listOfCoordinatesNeighbors).head()
df
```

```python
NeighborsStatesInDataFrame([1,6],listOfCoordinatesNeighbors,df).head()
```

## Modelos con muerte por enfermedad

```python
ranges = [[1,25,0.4],[25,64,0.4],[65,90,0.2]]
mr = [[1,14,0.0005],[15,69,0.02],[70,100,0.15]]
br = 2
ages_sys3 = agesMatrix(ranges,system)
ranges_dead = [[1,20,0.08],[21,45,0.15],[46,78,0.35],[79,100,0.8]]

SIRmodelWithDeathByDisease = models("sir_deathByDisease",0.2,0.5,system,Moore)
SIRmodelWithDeathByDisease.birthRate = br; SIRmodelWithDeathByDisease.probabilityOfDyingByAgeGroup = mr
SIRmodelWithDeathByDisease.deathFromDiseaseByAgeRange = ranges_dead
SIRmodelWithDeathByDisease.systemAges = ages_sys3; SIRmodelWithDeathByDisease.annualUnit = 365
```

```python
SIRmodelWithDeathByDiseaseApplied = SIRmodelWithDeathByDisease.basicModel(10)

df = pd.DataFrame()
x,y = system.shape
n = x*y

df["# Susceptibles"] = [int(n*m) for m in SIRmodelWithDeathByDiseaseApplied[1][0]]
df["# Infectados"] = [int(n*m) for m in SIRmodelWithDeathByDiseaseApplied[1][1]]
df["# Recuperados"] = [int(n*m) for m in SIRmodelWithDeathByDiseaseApplied[1][2]]
df["# Muertos"] = [int(n*m) for m in SIRmodelWithBirthsAndDeathsApplied[1][3]]
df["% Susceptibles"] = SIRmodelWithDeathByDiseaseApplied[1][0]
df["% Infectados"] = SIRmodelWithDeathByDiseaseApplied[1][1]
df["% Recuperados"] = SIRmodelWithDeathByDiseaseApplied[1][2]
df["% Muertos"] = SIRmodelWithBirthsAndDeathsApplied[1][3]

df.head()
```

```python
SIRmodelWithDeathByDiseaseEvolutions = SIRmodelWithDeathByDisease.evolutions
listOfCoordinatesNeighbors = Neighbors(system,Moore,1,1)

df = GetStatesByIteration(SIRmodelWithDeathByDiseaseEvolutions,listOfCoordinatesNeighbors).head()
df
```
